- HashMap:

  Un mapa hash es una estructura de datos que asocia llaves o claves con ciertos valores. La operación principal que soporta de manera eficiente es la búsqueda: permite el acceso a los elementos almacenados a partir de una clave generada. Funciona transformando la clave con una función hash en un hash, un número que identifica la posición (casilla o cubeta) donde la tabla hash localiza el valor deseado. Esto lo hace al tener una lista de n posiciones, donde cada una va a tener asi mismo un arreglo dinámico, y entonces al momento de tener el número de algún objeto mediante hashing, aplica la operación módulo a n (resultando entre 0 a n-1), asignando la posición en la que se va a agregar el registro, agregandose en un arreglo dinámico en dicha posición, para así ser capaces de distinguir entre llaves, al momento de una posible colisión.
 
  Una colisión corresponde a una situación que se produce cuando dos entradas distintas a una función de hash producen la misma salida. Por ello, la generación de los códigos hashing, debe ser lo más elaborada posible, para que se pueda asegurar en lo mayor posible, las pocas colisiones de estas, y que preferiblemente cuenten con orden O (1), para que sea lo más eficiente posible.

  Una colision puede  suceder frecuentemente en funciones como 
  
  $h(x)=x (mod 7)$
   
  $h(9)=2$
    
  $h(2)=2$

  Al momento de pensar en colisiones, debemos tener en cuenta que tan probables es no tenerlas. Para p datos con N posibles casillas tendremos:

  $q(p, N) = \frac{N}{N} * \frac{N-1}{N} ... \frac{N-p+1}{N} =  \frac{(N-1)(N-1) ... (N-p+1)}{N^{p-1}}$

  Por lo que para p >= 23, ya tendremos una probabilidad de colisionar > 0.5. Esto quiere decir que no podremos evitar al 100% las colisiones, ya que la memoria no es infinita.

  Tambien hay que tener en cuenta que, un buen N a escoger para un HashMap sería algún número primo. El problema con los no primos, es que al momento de obtener el código hash, es más probable que este resulte múltiplo o tenga factores comúnes en la descomposición en terminos de sus factores primos, corriendo el riesgo de que terminen en alguna casilla repetida.

  Supongamos que la función hash es:

  $h(x) = x (mod N)$

  Asumimos que las llaves están distribuidas en equidistancia:

  $\Delta < N$

  $k_{i} = z + i * \Delta$

  Obtendremos una colisión si:

  $k_{i} (mod N) = k_{j} (mod N)$

  $z + i * \Delta (mod N) = z + j * \Delta (mod N)$

  $i = j+m * N, m \in \mathbb{Z} $

  Para la busqueda, la entrada será un objeto cualquiera, obtenemos su código hash, aplicamos módulo de n, y así sabremos en que posición buscar, sin necesidad de recorrer toda la estructura, lo cual en tiempo de ejecución es muy eficiente. Una vez dirigidos a cierta posición, vamos a recorrer el arreglo dinámico que se encuentra en ella, buscando cual coincide con el objeto o llave ingresada, para retornar así el valor de esta.

  La eficiencia del hashing estará determinada por la función de hash, el tipo de llaves que se usen, distribución de estas, ocupancia del hashmap, y el método para manejar las colisiones.

  El valor de carga α corresponde al número de elementos en la tabla (n), dividido por el tamaño de esta (N):

  $\frac{n}{N}$

  Un α > 0.85 tiene efectos negativos en la eficiencia, debido a la sobrecarga de colisiones, por lo que debemos encontrar un punto medio entre llena y no muy llena.

- Cifrador RSA:

  RSA es un cifrador asimétrico, es decir que trabaja con dos claves, una pública y una privada. Todo el contenido de texto plano, o contenido sin cifrar, que sea hecho con la clave pública, podrá ser descifrado mediante la clave privada, y viceversa, todo contenido cifrado con la clave privada podrá ser descifrado mediante la clave pública. Para la generación de números primos se verifica que cada número no sea divisible por ninguno de los primos menores a su raíz cuadrada, y de no serlo, es agregado a la lista de primos para el posterior análisis. 

  El sistema RSA es una de las mayores aplicaciones en criptografía de la aritmetica modular, gracias a esta se consigue cifrar informacion de manera simple computacional y que solo quien conozca la llave para decifrar consiga hacerlo de manera igualmente sencilla. Gracias a esto se logra simplificar acciones computacionales o complicarlas a grandes escalas para mantener la
  
  El sistema realiza los siguientes pasos:

  1. Generar dos números primos muy grandes, p y q
  2. Hacer n=p*q
  3. Calcular z=(p-1)*(q-1)

  Se aplica la función fi Ⲫ de Euler para el caso de números primos,  y la propiedad multiplicativa de esta misma para números coprimos. 
  4. Elegir un pequeño número k, co-primo de z, de modo que el Máximo Común Divisor entre z y k sea 1, con 1<k<z
  5. Encontrar un número j tal que el (j mod z) sea 1
  6. Publicar k y n como la clave pública.
  7. Guardar j como la clave privada.

  - Para cifrar un mensaje:

      $P^k = E ( mod n )$

      P es el mensaje en texto plano

      n y k son la clave pública

      E es el mensaje cifrado

  - Para descifrar un mensaje:

      $E^j = P (mod n)$

      E es el mensaje cifrado

      j la clave privada

      P el mensaje en texto plano

      n es parte de la clave pública del destinatario

  El sistema RSA funciona basado en:

    $j*k=1+x(p-1)(q-1)$

    $E^j(mod n)=(P^k)^j (mod n)$

    $=P^(1+x(p-1)(q-1))(mod n) =(P^(p-1)(q-1))^x P(mod n)$

  Retomando la funcion fi de Euler: 

    $Ⲫ(n)=(p-1)(q-1)$

  Con el pequeño teorema de Fermat: 

    $P^(Ⲫ(n))=1(mod n) =>(P^(p-1)(q-1))^x=1(mod n)$

  Se obtiene: 

    $E^j(mod n)=P(mod n)=P$

El cifrado y descifrado son conmutativos, siempre y cuando se cifre con clave publica y descifre con la clave privada o viceversa.

```
import base64
import random
from time import time
from PIL import Image
from io import BytesIO


'''Algunas funciones que vamos a usar como máximo común divisor'''

def gcd(x, y):

    while x!=y:
        if x>y:
            x = x - y
        else:
            y = y - x

    return x
```

***Clase HashMap:***

Esta estructura que implementamos consta de 3 funciones:

- Set(key, value) permite ingresar una nueva llave con su valor, si la llave no existe dentro de esta, de lo contrario se cambiara el valor de dicha llave por el que hallamos ingresado.

- HasKey(key) permite verificar si una llave ya se encuentra dentro de la estructura.

- Get(key) nos retorna el valor de la llave ingresada, siempre y cuando se encuentre dentro de esta.

```
''' La clase Info que permite guardar una llave y su valor, y va a ser auxiliar
en la clase HashMap'''

class Info:
    
    def __init__(self, k, val):
        self.key = k
        self.value = val

''' La clase HashMap, donde como llave podemos ingresar un entero, string o 
cualquier tipo de dato que posea una función de hash, y como valor también 
podemos agregar cualquier otra cosa'''

class HashMap:

    def __init__(self, n):
        self.num = n
        self.lista = [None, ] * n

    def Set(self, k, val):
        
        ref = hash(k) % self.num

        if self.lista[ref] != None:

            for i in range(len(self.lista[ref])):
                if self.lista[ref][i].key == k:
                    self.lista[ref][i].value = val
                    return
            
            info = Info(k, val)
            self.lista[ref].append(info)

        else:
            self.lista[ref] = []
            info = Info(k, val)
            self.lista[ref].append(info)

    def HasKey(self, k):

        ref = hash(k) % self.num

        if self.lista[ref] != None:

            for i in range(len(self.lista[ref])):
                if self.lista[ref][i].key == k:
                    return True
        
        return False

    def Get(self, k):

        ref = hash(k) % self.num

        if self.lista[ref] != None:

            for i in range(len(self.lista[ref])):
                if self.lista[ref][i].key == k:
                    return self.lista[ref][i].value
            
        return None
```

Para transformar la clave en el numero identificador con la funcion hash se realiza una serie de operaciones o algoritmo matematico. Buscando la simplicidad en la ejecucion, se usa una expresion de sencilla y rapida. El uso de los numeros primos permite reducir el numero de posibles colisiones cuando se aplique el modulo.

```
'''Clase usuario creada para pruebas con el hashMap, para esta clase vamos a
crear nuestra propia función de hash, con el último método de esta'''

class User:
    
    def __init__(self, id, nombre, tipo):
        self.id = id
        self.nombre = nombre
        self.tipo = tipo

    def __eq__(self, other):
        return self.id == other.id and self.nombre == self.nombre

    def __hash__(self):
        sum = 0

        for i in range(len(self.nombre)):
            sum += ord(self.nombre[i])**(31-1-i)
        sum += self.id

        return sum
```

***Cifrador RSA:***

Para este vamos a tener dos opciones, una en donde el usuario ingresa los dos números primos para construir la clave privada y pública, o generarla aleatoriamente. Esto se consigue asignando una posición aleatoria, con un valor lo más alto que se pueda, para generar el número primo que se encuentra en dicha posición.

La llave pública y privada se generan haciendo las operaciones explicadas anteriormente.

```
'''Vamos a tener un RSA, que permite seleccionar dos números primos 
(y detectar cuando no lo son), o generarlos aleatoriamente. De esta forma, 
podemos generar la llave privada dentro de la clase con la cual cifraremos 
y descifraremos'''

class CifradorRSA:
    
    def __init__(self):

        self.__list_primes = [2, 3, 5, 7, 11, 13]
        self.__p = None
        self.__q = None
        self.n = None
        self.__z = None
        self.k = None
        self.__private_key = None

    def generate_random(self):
        self.__p = self.num_prime(random.randrange(100, 200))
        self.__q = self.num_prime(random.randrange(100, 200))
        self.n = self.__p * self.__q
        self.__z = (self.__p - 1) * (self.__q - 1)
        self.__select_k()
        self.__select_priv_key()

    def Select(self, p, q):
        self.__p = p
        self.__q = q

        if self.isprime(p) and self.isprime(q):
            self.__p = p
            self.__q = q
            self.n = self.__p * self.__q
            self.__z = (self.__p - 1) * (self.__q - 1)
            self.__select_k()
            self.__select_priv_key()

        else:
            raise Exception("Numbers in Select method must be primes")

    def __select_k(self):

        for i in range(1, len(self.__list_primes)):

            if gcd(self.__z, self.__list_primes[i]) == 1:
                self.k = self.__list_primes[i]
                break
    
    def __select_priv_key(self):

        for i in range(1, self.__z+1):
            if self.k*i % self.__z == 1:
                self.__private_key = i
                break

    def cifre(self, M):

        return (M**(self.k)) % self.n

    def descifre(self, E):

        return (E**(self.__private_key)) % self.n

    def isprime(self, n):

        if self.__list_primes[-1] >= n:
            if n in self.__list_primes:
                return True
        
        else:

            x = self.__list_primes[-1]+2
            z = 1

            while self.__list_primes[-1]<=n:
                if self.__list_primes[z] > x**(1/2):
                    self.__list_primes.append(x)
                    z = 1
                    x = x+2
                elif x%self.__list_primes[z] != 0:
                    z = z+1
                else:
                    x = x+2
                    z = 1

            if n in self.__list_primes:
                return True

        return False


    def num_prime(self, pos):

        x = self.__list_primes[-1]+2
        z = 1

        while len(self.__list_primes)<pos:

            if self.__list_primes[z] > x**(1/2):
                self.__list_primes.append(x)
                z = 1
                x = x+2
            elif x%self.__list_primes[z] != 0:
                z = z+1
            else:
                x = x+2
                z = 1
        
        return self.__list_primes[pos-1]

    def cifre_message(self, message):

        cif_mes = ""

        for i in range(len(message)):

            if i == len(message)-1:
                cif_mes += str(self.cifre(ord(message[i])))
            else:
                cif_mes += str(self.cifre(ord(message[i]))) + " "

        return cif_mes

    def descifre_message(self, message):

        orig_mes = ""
        message = message.split()

        for i in range(len(message)):
            orig_mes += chr(self.descifre(int(message[i])))

        return orig_mes
```
